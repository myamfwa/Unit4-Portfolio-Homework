# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:
1. whale_returns.csv
2. algo_returns.csv
3. sp500_history.csv

## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
# Reading whale returns
whale_returns_csv = Path("Resources/whale_returns.csv")

# Read in data and index by Date
whale_returns_dataframe = pd.read_csv(whale_returns_csv, index_col="Date", parse_dates=True, infer_datetime_format=True)
whale_returns_dataframe.sort_index(inplace = True)
whale_returns_dataframe.head()
```

```python
# Count nulls
whale_returns_dataframe.isnull().sum()
```

```python
# Drop nulls
whale_returns_dataframe = whale_returns_dataframe.dropna().copy()
whale_returns_dataframe
```

```python
#check whale_returns
whale_returns_dataframe.isnull().sum()
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns
algo_returns_csv = Path("Resources/algo_returns.csv")

# Read in data and index by Date
algo_returns_dataframe = pd.read_csv(algo_returns_csv, index_col="Date", parse_dates=True, infer_datetime_format=True)
algo_returns_dataframe.sort_index(inplace = True)
algo_returns_dataframe.head()
```

```python
# Count nulls
algo_returns_dataframe.isnull().sum()
```

```python
# Drop nulls
algo_returns_dataframe = algo_returns_dataframe.dropna().copy()
algo_returns_dataframe
```

## S&P 500 Returns

Read the S&P500 Historic Closing Prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices
sp500_history_csv = Path("Resources/sp500_history.csv")
# Read in data and index by Date
sp500_history_dataframe = pd.read_csv(sp500_history_csv, index_col="Date", parse_dates=True, infer_datetime_format=True)
sp500_history_dataframe.sort_index(inplace = True)
sp500_history_dataframe.head()
```

```python
# Check Data Types
sp500_history_dataframe.dtypes
```

```python
# Clean identified numeric fields with $ symbol
sp500_history_dataframe["Close"] = sp500_history_dataframe["Close"].str.replace("$", "")
sp500_history_dataframe["Close"]
```

```python
# Fix Data Types
# Convert `order_total` from `object` to `float`
sp500_history_dataframe["Close"] = sp500_history_dataframe["Close"].astype("float")
```

```python
# Check Data Types
sp500_history_dataframe.dtypes
```

```python
#change to ascending order
sp500_history_dataframe.sort_index()
```

```python
# Calculate Daily Returns & Drop nulls
sp500_returns_daily = sp500_history_dataframe.pct_change().dropna().copy()
sp500_returns_daily.head()
```

```python
#check sp_500_history_dataframe
sp500_history_dataframe.isnull().sum()
```

```python
# Rename Column
sp500_returns_daily = sp500_returns_daily.rename(columns={
    "Close": "S&P 500"
})
sp500_returns_daily.head()
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Concatenate all DataFrames into a single DataFrame
joined_portfolio = pd.concat(
    [whale_returns_dataframe, algo_returns_dataframe, sp500_returns_daily], axis="columns", join="inner"
)
joined_portfolio.head()
```

# Portfolio Analysis

In this section, you will calculate and visualize performance and risk metrics for the portfolios.

## Performance

Calculate and Plot the daily returns and cumulative returns. Does any portfolio outperform the S&P 500? 

```python
# Plot daily returns
joined_portfolio.plot(figsize=(20,10))
```

```python
# Plot cumulative returns
joined_portfolio_cumulative_returns = (1 + joined_portfolio).cumprod()
joined_portfolio_cumulative_returns.plot(figsize=(20,10))
```

## Risk

Determine the _risk_ of each portfolio:

1. Create a box plot for each portfolio. 
2. Calculate the standard deviation for all portfolios
4. Determine which portfolios are riskier than the S&P 500
5. Calculate the Annualized Standard Deviation

```python
# Box plot to visually show risk
joined_portfolio.plot.box(figsize=(20,10))
```

```python
# Daily Standard Deviations
# Calculate the standard deviation for each portfolio. 
# Which portfolios are riskier than the S&P 500?
daily_std = joined_portfolio.std()
print(daily_std)
```

```python
# Determine which portfolios are riskier than the S&P 500
risk_daily_std = daily_std.copy()

for x in risk_daily_std.keys():
    if risk_daily_std[x] <= daily_std["S&P 500"]:
        risk_daily_std[x]="False"
    else:
        risk_daily_std[x]="True"
print(risk_daily_std)
```

```python
# Calculate the annualized standard deviation (252 trading days)
annualized_std = daily_std * np.sqrt(252)
print(annualized_std)
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Plot the rolling standard deviation of the various portfolios along with the rolling standard deviation of the S&P 500 (consider a 21 day window). Does the risk increase for each of the portfolios at the same time risk increases in the S&P?
2. Construct a correlation table for the algorithmic, whale, and S&P 500 returns. Which returns most closely mimic the S&P?
3. Choose one portfolio and plot a rolling beta between that portfolio's returns and S&P 500 returns. Does the portfolio seem sensitive to movements in the S&P 500?
4. An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the ewm with a 21 day half-life.

```python
# Calculate and plot the rolling standard deviation for
# the S&P 500 and whale portfolios using a 21 trading day window
joined_portfolio.rolling(window=21).std().plot(figsize=(20,10))
```

```python
# Construct a correlation table
correlation = joined_portfolio.corr()
correlation
```

```python
# Calculate Beta for a single portfolio compared to the total market (S&P 500) - PAULSON & CO.INC. & S&P500
# (Your graph may differ, dependent upon which portfolio you are comparing)
PAULSON_SP500 = joined_portfolio.drop(columns=
                                      ["SOROS FUND MANAGEMENT LLC", "TIGER GLOBAL MANAGEMENT LLC", "BERKSHIRE HATHAWAY INC", "Algo 1", "Algo 2"]
                                     )
PAULSON_SP500.head()

rolling_variance = PAULSON_SP500['S&P 500'].rolling(window=60).var()
rolling_covariance = PAULSON_SP500['PAULSON & CO.INC. '].rolling(window=60).cov(PAULSON_SP500['S&P 500'])
rolling_beta = rolling_covariance / rolling_variance
rolling_beta.plot(figsize=(20, 10), title='Rolling 60-Day Beta of PAULSON & CO. INC.')
```

```python
# Calculate a rolling window using the exponentially weighted moving average. 
joined_portfolio.rolling(window=60).mean().plot(figsize=(20,10))
# joined_portfolio.ewm(adjust=True).mean().plot(figsize=(20,10))
```

## Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. (After all, if you could invest in one of two portfolios, each offered the same 10% return, yet one offered lower risk, you'd take that one, right?)

1. Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot.
2. Determine whether the algorithmic strategies outperform both the market (S&P 500) and the whales portfolios.

```python
# Calculate annualized Sharpe Ratios
sharpe_ratios = (joined_portfolio.mean() * 252) / (joined_portfolio.std() * np.sqrt(252))
sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios.plot(kind="bar", title="Sharpe Ratios")
```

## Choose 3-5 custom stocks with at last 1 year's worth of historic prices and create a DataFrame of the closing prices and dates for each stock.

```python
# Read the first stock - Netflix

nflx_historical_csv = Path("Resources/nflx_historical.csv")

# Read in data and index by Date
nflx_historical_dataframe = pd.read_csv(nflx_historical_csv, index_col="Date", parse_dates=True, infer_datetime_format=True)
nflx_historical_dataframe.sort_index(inplace = True)
nflx_historical_dataframe.index = nflx_historical_dataframe.index.date
nflx_historical_dataframe = nflx_historical_dataframe.rename(columns={"":"Date",
    "Close": "NFLX"
})
nflx_historical_dataframe.head()
```

```python
# Read the second stock - Sony
sne_historical_csv = Path("Resources/sne_historical.csv")

# Read in data and index by Date
sne_historical_dataframe = pd.read_csv(sne_historical_csv, index_col="Date", parse_dates=True, infer_datetime_format=True)
sne_historical_dataframe.sort_index(inplace = True)
sne_historical_dataframe.index = sne_historical_dataframe.index.date
sne_historical_dataframe = sne_historical_dataframe.rename(columns={
    "Close": "SNE"
})
sne_historical_dataframe.head()
```

```python
# Read the second stock - Facebook
fb_historical_csv = Path("Resources/fb_historical.csv")

# Read in data and index by Date
fb_historical_dataframe = pd.read_csv(fb_historical_csv, index_col="Date", parse_dates=True, infer_datetime_format=True)
fb_historical_dataframe.sort_index(inplace = True)
fb_historical_dataframe.index = fb_historical_dataframe.index.date
fb_historical_dataframe = fb_historical_dataframe.rename(columns={
    "Close": "FB"
})
fb_historical_dataframe.head()
```

```python
# Concatenate all stocks into a single DataFrame
musonda_portfolio = pd.concat(
    [nflx_historical_dataframe, sne_historical_dataframe, fb_historical_dataframe], axis="columns", join="inner"
)
musonda_portfolio.head()
```

```python
# Reset the index
musonda_portfolio.sort_index(inplace = True)
musonda_portfolio.head()
```

```python
# Pivot so that each column of prices represents a unique symbol - completed above
```

```python
# Drop Nulls
musonda_portfolio_daily_returns = musonda_portfolio.pct_change().dropna().copy()
musonda_portfolio_daily_returns.head()
```

```python
#Check for Nulls
musonda_portfolio_daily_returns.isnull().sum()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Calculate weighted portfolio returns
weights = [1/3, 1/3, 1/3]
musonda_portfolio_returns = musonda_portfolio_daily_returns.dot(weights)
musonda_portfolio_returns.head()
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# Add your "Custom" portfolio to the larger dataframe of fund returns
all_joined_portfolio = pd.concat(
    [joined_portfolio, musonda_portfolio_returns], axis="columns", join="inner"
)
columns = ["SOROS FUND MANAGEMENT LCC", "PAULSON & CO.INC. ", "TIGER GLOBAL MANAGEMENT LLC", "BERKSHIRE HATHAWAY INC", "Algo 1", "Algo 2", "S&P 500", "custom_musonda"]
all_joined_portfolio.columns = columns
all_joined_portfolio.head()
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)
all_joined_portfolio = all_joined_portfolio.dropna().copy()
all_joined_portfolio.head()
```

## Re-run the performance and risk analysis with your portfolio to see how it compares to the others

```python
# Risk
daily_std = all_joined_portfolio.std()
print(daily_std)
```

```python
#Annualized risk
annualized_std = daily_std * np.sqrt(252)
print(annualized_std)
```

```python
# Rolling - 21 Days
all_joined_portfolio.rolling(window=21).std().plot(title="21 Day Rolling Standard Deviation", figsize=(20,10))
```

```python
# Annualized Sharpe Ratios
sharpe_ratios = (all_joined_portfolio.mean() * 252) / (all_joined_portfolio.std() * np.sqrt(252))
sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios.plot(kind="bar", title="Sharpe Ratios")
```

```python
# Create a correlation analysis
correlation = all_joined_portfolio.corr()
correlation
```

```python
# Beta
custom_musonda = all_joined_portfolio.drop(columns=
                                      ["SOROS FUND MANAGEMENT LCC", "PAULSON & CO.INC. ", "TIGER GLOBAL MANAGEMENT LLC", "BERKSHIRE HATHAWAY INC", "Algo 1", "Algo 2"]
                                     )
custom_musonda.head()

rolling_variance = custom_musonda['S&P 500'].rolling(window=60).var()
rolling_covariance = custom_musonda['custom_musonda'].rolling(window=60).cov(custom_musonda['S&P 500'])
rolling_beta = rolling_covariance / rolling_variance
rolling_beta.plot(figsize=(20, 10), title='Rolling 60-Day Beta of musonda_custom')
```
